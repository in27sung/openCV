## 이미지 저장

### 저장 포맷(jpg)

```python
import cv2
img = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE) # 흑백으로 이미지 불러오기

while True:
    cv2.imshow('img', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cv2.waitKey(1)

result = cv2.imwrite('img_save.jpg', img) # jpg 형태로 이미지 저장 
print(result)
```

### 저장 포맷(png)

```python
import cv2
img = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE) # 흑백으로 이미지 불러오기
cv2.imwrite('img_save.png', img) #png 형태로 이미지 저장
```

## 동영상 저장 

```python
import cv2
cap = cv2.VideoCapture('video.mp4')

# 코덱 정의(fourCharacterCode) 
fourcc = cv2.VideoWriter_fourcc(*'H264') # == ('H', '2', '6', '4')

# 프레임 크기, FPS
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # datetype(int)
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # datetype(int)
fps = cap.get(cv2.CAP_PROP_FPS) * 2 # 재생속도 2배

out = cv2.VideoWriter('output_fps(*2).mov', fourcc, fps, (width, height))
# 저장 파일명, 코덱, FPS, 크기(width, height)

while cap.isOpened():
    ret, frame = cap.read() # 성공시 true 
    
    if not ret:
        break

    out.write(frame) # 영상 데이터만 저장 (소리x)        
    cv2.imshow('video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

out.release() # 자원 해제        
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
